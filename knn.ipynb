```python
class Knn:
    def __init__(self, k):
        self.k = k
        self.dataset = []
    def fit(self, X, Y):
        self.dataset = list(zip(X,Y))
        #for x,y in zip(X,Y):
            #self.neighbors.append(x, y)
    def get_nearest_neighbors(self, a):
        nearest_neighbors = sorted(self.dataset, key = lambda b: self.get_distance(a, np.array(b[0])))[:self.k]
        return nearest_neighbors
    def get_distance(self, a, b):
        distance = np.sqrt(np.sum((a-b)**2))
        return distance
    def predict(self, x):
        nearest_neighbors = self.get_nearest_neighbors(x)
        values = list(zip(*nearest_neighbors))[1]
        Y = Counter(values).most_common(1)[0][0]
        return Y
    def score(self, X, Y):
        correct = 0
        total = len(Y)
        for x,y in zip(X,Y):
            if self.predict(x)==y:
                correct += 1
        score = correct/total
        return score
     
        
X, Y = load_iris(return_X_y = True)
X_train, X_test, Y_train, Y_test = train_test_split(X,Y)
model = Knn(3)
model.fit(X_train,Y_train)
X_= list(zip(*X))
colours = {0:"g",1:"b",2:"r"}
plt.scatter(X_[0],X_[1],c=[colours[y] for y in Y])
model.score(X_test, Y_test)


```
